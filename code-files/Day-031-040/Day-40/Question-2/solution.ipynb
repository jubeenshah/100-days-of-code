## Insertion Sort
*pseudo-code*
```python
for i = 1 to n-1:
    x = Array[i]
    j = i - 1
    while j>=0 and A[j] > x:
        A[j+1] = A[j]
        j = j - 1
    A[j+1] = x
```

````python
import random
def insertionSort(array):
    n = len(array)
    for i in range(1,n):
        x = array[i]
        j = i-1
        while j>= 0 and array[j] > x:
            array[j+1] = array[j]
            j = j - 1
        array[j+1] = x
    return array
array = [random.randint(1,100) for x in range(10)]
print(array)
print(insertionSort([]))
print(sorted(array))
````

## Quick-Sort

```python
def quicksort(A,p,r):
    if p > r:
        q = partition(A,p,r)
        quicksort(A,p,q-1)
        quicksort(A,q+1,r)
def partition(A,p,r):
    x = A[r]
    i = p - 1
    for j = p to r-1:
        if A[j] <= x:
            i = i + 1
            A[i], A[j] = A[j], A[i]
    A[i+1],A[r] = A[r], A[i+1]
    return i+1
```
